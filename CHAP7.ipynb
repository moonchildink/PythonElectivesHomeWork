正则表达式及re模块

```python
import re
testStr = 'H(efei'
s= re.match('\w{1,3}', testStr)
print(s)
```

案例

```python
#匹配手机
testStr = '19456789681'
pat = '\d{1}[34578]{1}\d{9}'

res = re.match(pat, testStr)
print(res)
```

```python
#验证邮箱
testStr = 'xsl3457@163.com'
pat = '\w+@[a-z0-9]+\.[a-z]{1,3}'
res =re.match(pat, testStr)
print(res)
```

Search方法及位置数量元字符

```python
testStr = 'Hello world! world!'
res = re.search('world!', testStr)
print(res)
res
```

贪婪及非贪婪模式

```python
testStr = '123456789'
pat = '\d+?'
res = re.search(pat, testStr)
res.group()
```

```python
testStr = '<h1>This is a title</h1>'
pat = '<.+?>'
res =  re.search(pat, testStr)
res.group()
```

正则表达式中的转义字符

```python
testStr = r'The apple price is $99, \the orange price is $88'
pat = '\$\d+'
res = re.search(pat, testStr)
print(res)
```

分组（）

```python
testStr = r'The apple price is $99, \the orange price is $88'
pat = '.+(\$\d+).+(\$\d+)'
res = re.search(pat, testStr)
res.group(2)
```

```python
testStr = '<html><head>This is head.</head><body>This is body.</body></html>'
pat = re.compile(r'<(\w+)><(\w+)>(.+)</\2><(\w+)>(.+)</\4></\1>')
res = re.findall(pat, testStr)
res
```

```python
#findalld方法
res = re.findall('\$\d+', testStr )
res
```
